```python
import pandas as pd 
```

```python
data = pd.read_csv("aluguel.csv", sep = ';')
data.head()
```

```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'.
list(data['Tipo'].drop_duplicates())
```

```python
selec_apartment = ['Apartamento']
selec_apartment_data = data['Tipo'].isin(selec_apartment)
selec_apartment_data
```

```python
apartament = data[selec_apartment_data]
apartament
```

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
data['Tipo'].drop_duplicates()
```

```python
houses_data_list = ['Casa', 'Casa de Condomínio', 'Casa de vila']
houses_data_list_selec = data['Tipo'].isin(houses_data_list)
houses_data_list_selec
```

```python
houses = data[houses_data_list_selec]
houses
```

```python
houses.shape[0]
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
data.head()
```

```python
selecao = (data['Area'] >= 60) & (data['Area'] <= 100)
selecao
```

```python
area = data[selecao]
area
```

```python
area.shape[0]
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
data.head()
```

```python
selecao = (data['Quartos'] >= 4) & (data['Valor'] < 2000)
beedrooms =  data[selecao]
beedrooms
```

```python
beedrooms.shape[0]
```

```python
print('Selecione somente os imóveis classificados com tipo Apartamento foram {}'.format(apartament.shape[0]))
print('Selecione os imóveis classificados com tipos Casa, Casa de Condomínio e Casa de Vila, foram {}'.format(houses.shape[0]))
print('Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites, forma {}'.format(area.shape[0]))
print('Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00, foram {}'.format(beedrooms.shape[0]))
```
