```python
import pandas as pd
```

```python
dados = pd.read_csv("aluguel.csv", sep=';')
dados.head()
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencia_lista = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa Comercial',
 'Casa de Vila',
 'Loteamento/Condomínio',
 'Sítio',]
```

```python
selecao = dados['Tipo'].isin(residencia_lista)
selecao.head()
```

```python
dados[selecao]
```

```python
list(dados[selecao]['Tipo'].drop_duplicates())
```

```python
residencia = dados[selecao]
residencia.head()
```

```python
residencia.isnull()
```

```python
residencia.info()
```

```python
residencia[residencia['Valor'].isnull()]
```

```python

residencia.dropna(subset = ['Valor'], inplace = True)

```

```python
residencia[residencia['Valor'].isnull()]
```

```python
residencia.head()
```

```python
selecao = (residencia['Tipo'] == 'Apartamento') & (residencia['Condominio'].isnull())
```

```python
# botando o ~ se faz a inversao o que e true vira falso e o que e falso vira true
residencia = residencia[~selecao]
```

```python
residencia = residencia.fillna({'Condominio':0, 'IPTU': 0})
```

```python
residencia
```

```python
residencia.info()
```

```python
residencia.to_csv('residencia.csv')
```
