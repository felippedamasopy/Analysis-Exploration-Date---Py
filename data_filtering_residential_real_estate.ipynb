```python
import pandas as pd
```

```python
data = pd.read_csv('aluguel.csv', sep=';')
data.head()
```

```python
list(data['Tipo'].drop_duplicates())
```

```python
residential_realstate = ['Quitinete',
 'Casa',
 'Casa de Condomínio',
 'Casa de Vila',]
 
```

```python
 data['Tipo'].isin(residential_realstate)
```

```python
selection =  data['Tipo'].isin(residential_realstate)
selection
```

```python
data[selection]
```

```python
residential_real_state = data[selection]
residential_real_state
```

```python
residential_real_state.shape[0]
```

```python
residential_real_state.index = range(residential_real_state.shape[0])
residential_real_state
```

```python
residential_real_state.to_csv('residential_real_state2.csv', index= False)
```
